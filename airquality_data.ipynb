```
climateDF = spark.read.format('csv').options(
    header='true', inferschema='true').load("/mnt/data/*_bme280sof.csv")
pollutionDF = spark.read.format('csv').options(
    header='true', inferschema='true').load("/mnt/data/*_sds011sof.csv")
```

```
climateDF = climateDF.drop('_c0').drop('sensor_id').drop('lat').drop('lon')
```

```
air_qualityDF = pollutionDF.join(climateDF, how='inner', on=['timestamp','location'])
```

```
air_qualityDF.printSchema()
```

```
air_qualityDF.show()
```

```
air_qualityDF = air_qualityDF.drop('_c0', 'sensor_id', 'location')
```

```
air_quality_shape = (air_qualityDF.count(), len(air_qualityDF.columns))
print(air_quality_shape)
```

```
from pyspark.sql.functions import to_date, dayofmonth, month, year, hour

air_qualityDF = air_qualityDF.withColumn('date', to_date('timestamp')).withColumn('hour', hour('timestamp'))
```

```
air_qualityDF = air_qualityDF.withColumn('year', year('date')).withColumn('month', month('date')).withColumn('day', dayofmonth('date'))
air_qualityDF = air_qualityDF.drop('timestamp')
```

```
air_qualityDF = air_qualityDF.drop('date')
air_qualityDF = air_qualityDF.na.drop()
air_qualityDF.show()
```

```
required_features = ['lat', 'lon', 'year', 'month', 'day', 'hour', 'pressure', 'temperature', 'humidity', 'P1']

from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(inputCols=required_features, outputCol='features')
transformDF = assembler.transform(air_qualityDF)
```

```
(trainDF, testDF) = transformDF.randomSplit([0.7, 0.3])
```

```
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(labelCol='P2', featuresCol='features')
model_lr = lr.fit(trainDF)
```

```
prediciones_lr = model_lr.transform(testDF)
```

```
from pyspark.ml.evaluation import RegressionEvaluator
evaluator = RegressionEvaluator(labelCol='P2', predictionCol='prediction')
r2 = evaluator.evaluate(prediciones_lr, {evaluator.metricName: "r2"})
rmse = evaluator.evaluate(prediciones_lr, {evaluator.metricName: "rmse"})
```

```
print("R2: {0}".format(r2))
print("RMSE: {0}".format(rmse))
```

```
from pyspark.ml.regression import DecisionTreeRegressor

dt = DecisionTreeRegressor(maxDepth=2, labelCol='P2', featuresCol='features')
model_dt = dt.fit(trainDF)
predicciones_dt = model_dt.transform(testDF)

r2 = evaluator.evaluate(predicciones_dt, {evaluator.metricName: "r2"})
rmse = evaluator.evaluate(predicciones_dt, {evaluator.metricName: "rmse"})
```

```
print("R2: {0}".format(r2))
print("RMSE: {0}".format(rmse))
```

```
from pyspark.ml.regression import GBTRegressor

gbt = GBTRegressor(maxDepth=2, seed=64, leafCol="leafId", labelCol='P2', featuresCol='features')
model_gbt = gbt.fit(trainDF)
predicciones_gbt = model_gbt.transform(testDF)

r2 = evaluator.evaluate(predicciones_gbt, {evaluator.metricName: "r2"})
rmse = evaluator.evaluate(predicciones_gbt, {evaluator.metricName: "rmse"})
```

```
print("R2: {0}".format(r2))
print("RMSE: {0}".format(rmse))
```
